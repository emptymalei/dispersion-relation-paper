## Blob Plots of Dispersion Relations

A simple transformation that turns the color bar plots into blob plots.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mpl
from matplotlib import rcParams

mpl.rcParams['figure.figsize'] = (12,12*0.618)

rcParams['font.family'] = 'sans-serif'
#rcParams['font.sans-serif'] = ['Computer Modern Roman']
# Times-Roman

ftsz = 22
lsaPTS = 30
CBAR = 'viridis_r'
```

```python
plt.style.use('seaborn-ticks') #seaborn-ticks
```

### Discrete Zenith Angles

#### 2 Angles

```python
%%bash
ls data/dr-db-two
```

```python
# Import data for WC1 spectrum
spectDBWC1DRDBPltDataUpper = np.transpose( np.genfromtxt ('data/dr-db-two/spectDBWC1DRDBPltData-upper.csv', delimiter=",")  )
spectDBWC1DRDBPltDataLower = np.transpose( np.genfromtxt ('data/dr-db-two/spectDBWC1DRDBPltData-lower.csv', delimiter=",")  )

spectDBWC1DRDBMZAPltDataUpper = np.transpose( np.genfromtxt ('data/dr-db-two/spectDBWC1DRDBMZAPltData-upper.csv', delimiter=",")  )
spectDBWC1DRDBMZAPltDataLower = np.transpose( np.genfromtxt ('data/dr-db-two/spectDBWC1DRDBMZAPltData-lower.csv', delimiter=",")  )

lsaMAARODataDBWC1 = np.transpose( np.genfromtxt ('data/dr-db-two/lsaMAARODataDBWC1.csv', delimiter=",")  )
lsaMZApRODataDBWC1 = np.transpose( np.genfromtxt ('data/dr-db-two/lsaMZApRODataDBWC1.csv', delimiter=",")  )
lsaMZAmRODataDBWC1 = np.transpose( np.genfromtxt ('data/dr-db-two/lsaMZAmRODataDBWC1.csv', delimiter=",")  )

# Import data for C1 spectrum
spectDBC1DRDBPltDataUpper = np.transpose( np.genfromtxt ('data/dr-db-two/spectDBC1DRDBPltData-upper.csv', delimiter=",")  )
spectDBC1DRDBPltDataLower = np.transpose( np.genfromtxt ('data/dr-db-two/spectDBC1DRDBPltData-lower.csv', delimiter=",")  )

spectDBC1DRDBMZAPltDataUpper = np.transpose( np.genfromtxt ('data/dr-db-two/spectDBC1DRDBMZAPltData-upper.csv', delimiter=",")  )
spectDBC1DRDBMZAPltDataLower = np.transpose( np.genfromtxt ('data/dr-db-two/spectDBC1DRDBMZAPltData-lower.csv', delimiter=",")  )

lsaMAARODataDBC1 = np.transpose( np.genfromtxt ('data/dr-db-two/lsaMAARODataDBC1.csv', delimiter=",")  )
lsaMZApRODataDBC1 = np.transpose( np.genfromtxt ('data/dr-db-two/lsaMZApRODataDBC1.csv', delimiter=",")  )
lsaMZAmRODataDBC1 = np.transpose( np.genfromtxt ('data/dr-db-two/lsaMZAmRODataDBC1.csv', delimiter=",")  )
```

```python
# Generate the boundaries using Re(k)+Im(k) and Re(k)-Im(k)

lsaMZARODataDBWC1 = np.transpose( np.concatenate( (np.transpose( lsaMZApRODataDBWC1) , np.transpose( lsaMZAmRODataDBWC1 ) ), axis =0 )  )

# each element in lsaMZARODataDBWC1 denoted by ele, ele[1] is real part of k, ele[0] is omega, ele[2] is imaginary part of k.

lsaMAARODataDBWC1Blob = []

for ele in np.transpose( lsaMAARODataDBWC1 ):
    lsaMAARODataDBWC1Blob.append( [ele[0], ele[1], ele[1]+ele[2], ele[1]-ele[2] ] )

lsaMAARODataDBWC1Blob = np.transpose( np.array(lsaMAARODataDBWC1Blob) )

    
lsaMZARODataDBWC1Blob = []

for ele in np.transpose(lsaMZARODataDBWC1):
    lsaMZARODataDBWC1Blob.append( [ele[0], ele[1], ele[1]+ele[2], ele[1]-ele[2] ] )
    
lsaMZARODataDBWC1Blob = np.transpose( np.array(lsaMZARODataDBWC1Blob) )


```

```python
lsaMAARODataDBWC1.shape,\
lsaMAARODataDBWC1Blob.shape,\
lsaMZApRODataDBWC1.shape, \
lsaMZAmRODataDBWC1.shape, \
lsaMZARODataDBWC1.shape, \
lsaMZARODataDBWC1Blob.shape
```

```python
spectDBWC1DRDBMAAPlt,  spectDBWC1DRDBMAAPltax = plt.subplots()
plt.plot(spectDBWC1DRDBPltDataUpper[0],spectDBWC1DRDBPltDataUpper[1],'-k')
plt.plot(spectDBWC1DRDBPltDataLower[0],spectDBWC1DRDBPltDataLower[1],'-k')
db3c4seg =1
# lsaMAARODataDBWC1Plt = plt.scatter(lsaMAARODataDBWC1[1][::db3c4seg],lsaMAARODataDBWC1[0][::db3c4seg], c=lsaMAARODataDBWC1[2][::db3c4seg], s=lsaPTS, vmin=0, cmap=CBAR, edgecolors='none')
lsaMAARODataDBWC1Plt = plt.plot(lsaMAARODataDBWC1Blob[1][::db3c4seg],lsaMAARODataDBWC1Blob[0][::db3c4seg],'-r')
lsaMAARODataDBWC1PltP = plt.plot(lsaMAARODataDBWC1Blob[2][::db3c4seg],lsaMAARODataDBWC1Blob[0][::db3c4seg],'--r')
lsaMAARODataDBWC1PltM = plt.plot(lsaMAARODataDBWC1Blob[3][::db3c4seg],lsaMAARODataDBWC1Blob[0][::db3c4seg],'--r')
# plt.colorbar(lsaMAARODataDBWC1Plt)
plt.axhline(y=0, xmin=0, xmax=1, linewidth=1, color = 'k')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.xlabel('$k$')
plt.ylabel('$\omega$')
# plt.title('Dispersion Relation and Instabilities for WC1')
plt.savefig('../assets/spectDBWC1DRDBMAAPltBlob.pdf',bbox_inches='tight')
plt.show()
```

```python
spectDBWC1DRDBMZApPlt,  spectDBWC1DRDBMZApPltax = plt.subplots()
plt.plot(spectDBWC1DRDBMZAPltDataUpper[0],spectDBWC1DRDBMZAPltDataUpper[1],'-k')
plt.plot(spectDBWC1DRDBMZAPltDataLower[0],spectDBWC1DRDBMZAPltDataLower[1],'-k')
db3c4seg=1
# lsaMZApRODataDBWC1Plt =plt.scatter(lsaMZApRODataDBWC1[1][::db3c4seg],lsaMZApRODataDBWC1[0][::db3c4seg], c=lsaMZApRODataDBWC1[2][::db3c4seg], s=lsaPTS, vmin=0, cmap=CBAR, edgecolors='none')
# plt.colorbar(lsaMZApRODataDBWC1Plt)
lsaMZARODataDBWC1Plt =plt.plot( lsaMZARODataDBWC1Blob[1][::db3c4seg], lsaMZARODataDBWC1Blob[0][::db3c4seg] ,'-r' )
lsaMZARODataDBWC1PltP =plt.plot( lsaMZARODataDBWC1Blob[2][::db3c4seg], lsaMZARODataDBWC1Blob[0][::db3c4seg] ,'--r' )
lsaMZARODataDBWC1PltM =plt.plot( lsaMZARODataDBWC1Blob[3][::db3c4seg], lsaMZARODataDBWC1Blob[0][::db3c4seg] ,'--r' )
plt.axhline(y=0, xmin=0, xmax=1, linewidth=1, color = 'k')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.xlabel('$k$')
plt.ylabel('$\omega$')
# plt.title('Dispersion Relation and Instabilities for WC1')
plt.savefig('../assets/spectDBWC1DRDBMZAPltBlob.pdf',bbox_inches='tight')
plt.show()
```
